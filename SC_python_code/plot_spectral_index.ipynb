```python
import numpy as np
import pandas as pd
# Set up matplotlib and use a nicer set of plot parameters
import matplotlib
#matplotlib.rc_file("../../templates/matplotlibrc")
import matplotlib.pyplot as plt
matplotlib.rc('text', usetex=True)
import seaborn as sns
from matplotlib.colors import LogNorm
```

```python
spectral_index_nparray=np.empty([20,20])
```

```python
spectral_index_nparray = np.loadtxt('spectralindex.txt')
```

```python
print(np.abs(spectral_index_nparray))
```

```python
fig, ax = plt.subplots(figsize=(10,10)) 
cax=ax.matshow(np.abs(np.transpose(spectral_index_nparray)),cmap='plasma',origin='lower')
#print(chi_new.shape)
ax.tick_params(axis="x", bottom=True, top=False, labelbottom=True, labeltop=False)
ax.set_xticks([i for i in range(0,20)])
ax.set_xticklabels([str(i) for i in range(1,21)], fontsize=14)
#ax.set_yticklabels([Categories[i] for i in range(20)],fontsize=14)
#ax.colorbar()
ax.set_yticks([i for i in range(0,20)])
ax.set_yticklabels([str(i) for i in range(1,21)], fontsize=14)
ax.set_xlabel("Diffusion Coefficient "r'$10^{28}cm^{2}s^{-1}$', fontsize = 14)
ax.set_ylabel("Halo Size (kpc)", fontsize = 14)
fig.colorbar(cax)
plt.savefig("spectral_index_colormap.png",dpi=400)
plt.show()
```
